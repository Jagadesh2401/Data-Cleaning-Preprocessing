```python
df = pd.read_csv("medical-appointments-no-show.csv")
df.head()
df.info()
df.describe()
df.shape
```


# HANDLE MISSING VALUES

```python
missing_before = df.isnull().sum()

print(missing_before)
```

```python
df_clean = df.dropna()
```

```python
missing_after = df_clean.isnull().sum()

print(missing_after)
```

# REMOVE DUPLICATES

```python
duplicates_before = df_clean.duplicated().sum()
df.shape
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
duplicates_after = df_clean.duplicated().sum()
df_clean.shape
```

# CLEAN COLUMN NAMES

```python
df_clean.columns = (
    df_clean.columns.str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

# CONVERT DATE COLUMNS

```python
date_cols = [col for col in df_clean.columns if "day" in col.lower()]
for col in date_cols:
    df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
```

#SAVE CLEANED FILE

```python
output_filename = "cleaned_medical_appointments.csv"
df_clean.to_csv(output_filename, index=False)
```

# Prepare summary

```python
summary = {
    "Original Shape": df.shape,
    "New Shape": df_clean.shape,
    "Missing Values Before": missing_before.to_dict(),
    "Missing Values After": missing_after.to_dict(),
    "Duplicates Before": duplicates_before,
    "Duplicates After": duplicates_after,
    "Date Columns Detected": date_cols,
}
```
